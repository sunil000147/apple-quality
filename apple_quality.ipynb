# Checking for outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i in df.columns:
  sns.boxplot(df[i])
  plt.title(i)
  plt.show()
```

```python
df.columns
```

```python
def outlier_handling(column):

  lower_fence=np.quantile(column,0.10)
  print('lower_fence=',lower_fence)
  higher_fence=np.quantile(column,0.90)
  print('higher_fence=',higher_fence)
  outliers=[]
  for i in column:
    if i>higher_fence or i<lower_fence:
      outliers.append(i)
  print('number of outliers=',len(outliers))
  print()
```

```python
for i in df[['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness',
       'Ripeness', 'Acidity']]:
  print(i)
  outlier_handling(df[i])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import  SVC
from sklearn.linear_model import LogisticRegression
```

```python
dtc=DecisionTreeClassifier()
rfc=RandomForestClassifier()
xgb=XGBClassifier()
lgbm=LGBMClassifier()
svc=SVC()
lr=LogisticRegression()
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score
```

```python
def modeling(model,x_train,y_train):
  model.fit(x_train,y_train)

  y_pred=model.predict(x_test)
  accuracy=accuracy_score(y_test,y_pred)
  print('accuracy_score=',accuracy)
  print('classification_report=\n',classification_report(y_test,y_pred))
```

```python
modeling(dtc,x_train,y_train)
```
